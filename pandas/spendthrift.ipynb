```python
df_1_year['Date'] = pd.to_datetime(df_1_year['Date'], format='%d/%m/%y')
```

```python
df_1_year['Value Dt'] = pd.to_datetime(df_1_year['Value Dt'], format='%d/%m/%y')
```

```python
for index, row in df_1_year.iterrows():
    if row['Date'] != row['Value Dt']:
        delta = row['Value Dt'] - row['Date']
        if delta.days > 1:
            print('Delta:{}, Narration: {}, Date: {}, Value Dt: {}'.format(
                delta,
                row['Narration'], 
                row['Date'], 
                row['Value Dt'],
            ))
```

```python
# Here we realised that Value Dt is at max one day greater than Date
# if not the same day.
# Hence, dropping column Value Dt
```

```python
df_1_year = df_1_year.drop(columns='Value Dt')
```
