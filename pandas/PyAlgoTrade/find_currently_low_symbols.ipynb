```python
SQLALCHEMY_DATABASE_URI = 'mysql+pymysql://admin:admin@192.168.64.2:3306/dc_admin_local'
```

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
db_engine = create_engine(
    SQLALCHEMY_DATABASE_URI, 
    pool_size=10, 
    max_overflow=20,
    )
db_connection = db_engine.connect()
```

```python
# read data into df

df = pd.read_sql('''
SELECT symbol,
       AVG((open+ CLOSE) / 2) AS avg_price,
       YEAR(TIMESTAMP)
FROM quant.stock_daily_data
WHERE symbol LIKE '%%TATAMOTORS%%'
AND   TIMESTAMP BETWEEN '2000-01-01' AND '2020-12-31'
GROUP BY 1,
         3
ORDER BY 3;
''', db_connection);

```
